```python
from datasets import load_dataset
dataset = load_dataset("heegyu/namuwiki-extracted")
print(dataset["train"][0])
```

```python
from pprint import pprint
print(len(dataset['train']))
print(type(dataset['train']))

dic = dataset['train'][5]
dic.pop('contributors')
dic.pop('namespace')

pprint(dic)
```

```python
import os, pickle
from tqdm import tqdm

def make_pickle(dataset):
    
    SAVE_DIR = "./namu_wiki_dump/"
    pickles = []
    k = len(dataset['train'])
    
    for idx in tqdm(range(k)):
        SAVE_PATH = os.path.join(SAVE_DIR, f"{idx}.pkl")
        
        if os.path.exists(SAVE_PATH):
            continue
        
        dic = dataset['train'][idx]
        dic.pop('contributors')
        dic.pop('namespace')
        
        with open(SAVE_PATH, 'wb') as f:
            pickles.append(SAVE_PATH)
            pickle.dump(dic, f)
            
    return pickles
            
pickles = make_pickle(dataset)
print(len(pickles))
```
